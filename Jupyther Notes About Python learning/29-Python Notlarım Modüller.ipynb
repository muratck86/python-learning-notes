# datetime Modülü

```python
from datetime import datetime
datetime.now()
```

```python
print(datetime.now()) #Bu kullanımda görüldüğü gibi zamanı mikrosaniyeye kadar alıyoruz.
```

```python
zaman=datetime.now()
print(zaman.year)
print(zaman.month)
print(zaman.day)
print(zaman.hour) #kullanımlarıyla zamanın istediğimiz boyutunu alabiliyoruz
```

```python
print("Tarih:", datetime.ctime(zaman)) #ctime metodu 
print("Tarih2:", datetime.date(zaman)) #date metodu bunu şöyle de kullanabiliriz: zaman.date() aynı sonucu verecektir.
print("Özelleştirilmiş Yıl: ", zaman.strftime("%Y")) #strftime en çok kullanılan metodlardan, parametresiz kullanılamaz.
print("Özelleştirilmiş Yıl2: ", datetime.strftime(zaman, "%Y")) #diğer kullanımı,aynı sonuç.
print("Özelleştirilmiş Ay: ", zaman.strftime("%B"))
print("Özelleştirilmiş Gün: ", zaman.strftime("%C"))
print("Özelleştirilmiş Tarih: ", zaman.strftime("%D"))
print("Özelleştirilmiş Saat: ", zaman.strftime("%X"))
print("Özelleştirilmiş Tarih Saat: ", datetime.strftime(zaman, "%D %X"))
```

```python
print("Tarih :",zaman.strftime("%C.%B.%Y %A")) #İngilizce, bunu Türkçeye çevirmek için de bir modül mevcut
```

# requests ve beautifulsoup modülleri

```python
import requests
from bs4 import BeautifulSoup
```

```python
response=requests.get("https://www.kariyer.net/is-ilanlari/#&kw=Antalya")
#bu kod ile siteyi çekmiş oluruz ve çektiğimiz veriyi response adında bir değişkene attık.
response
```

```python
#Şimdi de response değişkenindeki sitenin html içeriğini alalım ve html_icerik adında bir değişkene gönderelim
html_icerik=response.content
html_icerik
```

**Yukarıda görüldüğü üzere bu içeriğin işimize yaraması için biraz düzenlememiz gerekiyor, bundan sonraki adımları BeautifulSoup ile yapıyoruz:**

```python
corba=BeautifulSoup(html_icerik,"html.parser") #içeriğimizi parçalamak için içine attık ve corba değişkenine atadık
corba
```

**Şimdi daha düzenli bir şekle kavuştu ancak daha güzel bir düzenleme için .prettify() kullanacağız:**

```python
print(corba.prettify())
```

**Web sayfasındaki "a" etiketlerini (başka bir dökümana götüren bağlantılar) almak için find_all() metodunu kullanacağız:**

```python
for i in corba.find_all("a"):
    print(i)
    print("-----------------------------------------------------------------------------------------------------------")
```

**Bu "a" etiketlerinden sadece linkleri almak için ise:**

```python
for i in corba.find_all("a"):
    print(i.get("href")) #Dikkat sözlük yapısına çok benziyor...
    print("-----------------------------------------------------------------------------------------------------------")
```

**Yine "a" etiketinden linklerde üzerine tıklanan textleri (yani "buraya tıklayın" gibi yazılar) almak için ise:**

```python
for i in corba.find_all("a"):
    print(i.text)
    print("-----------------------------------------------------------------------------------------------------------")
```

**Bu mantığı devam ettirerek html kodları içindeki etiketleri kullanarak istediğimiz tipteki verileri elde ediyoruz. Örneğin Kariyer.net'teki her bir ilan div class="ilan" bloklarında bulunur:**

```python
for i in corba.find_all("div",{"class":"ilan"}):
    print(i)
    print("-----------------------------------------------------------------------------------------------------------")
```
