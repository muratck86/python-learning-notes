Yukarıda "a" kipi ve "r" kipi ile dosyada ayrı ayrı yaptığımız okuma ve yazma işlemini "a+" kipiyle tek seferde yapabiliriz.

```python
with open("tahsilatlar.txt","a+") as f:
    f.write("Yusuf Davaz: 2500\n")
    f.seek(0)
    print(f.read())
```

```python
with open("tahsilatlar.txt","a+") as f:
    f.writelines("deneme: 2500\n")
    f.seek(0)
    print(f.read())
```

## Dosyanın başına veri ekleme

Burada da dosyamızı "r" kipi ile açarsak yazma işi yapamayız, "w" ile açarsak içindeki bilgileri kaybederiz."x" kipi zaten dosya mevcut olduğu için açmaz, "a+" Kipinde açmayı deneyelim:

```python
with open("tahsilatlar.txt","a+") as f:
    f.seek(0)
    dosyaverisi=f.read()
    f.seek(0)
    f.write("Cuma Kaya: 1500 TL\n"+dosyaverisi)
    f.seek(0)
    print(f.read())
```

Görüldüğü gibi "a+" kipi ile burada başa eklemek istediğimiz  bilgi yine en sona eklendi ve dosyanın ilk verisi de sona tekrar kopyalandı, bu durum amaçladığımızdan uzak...

```python
with open("tahsilatlar.txt","r+") as f:#dosyamızı okuma + yazma modunda açalım
    dosyaverisi=(f.read())
    f.seek(0)
    f.write("Satılmış Satar: 550 TL\n"+dosyaverisi)
    f.seek(0)
    print(f.read())
```

## truncate() metodu

Bu metodun kelime anlamı "kıpmak, budamak" anlamlarına gelir. Medotun görevi dosya boyutunu kırparak küçültmek ya da büyütmektir. Eğer parametresiz kullanırsak dosyayı 0 bayt boyutuna küçültür, yani dosyayı"w" kipinde açmışız gibi bütün içeriğini siler. Eğer 100 bayt boyutunda bir dosya için parametre olarak 50 girersek 50 bayta kadar olan veriden sonrasını silerek dosyayı 50 bayta kırpmış olur. Eğer 150 girersek dosya boyutunu 150 bayt yapacak kadar 0 ekler.

```python
with open("tahsilatlar2.txt","r+") as dosya: #örnek dosyamızın içinde ne varmış bakalım.
    print(dosya.read())
```

```python
#Bu örnek dosyamızdaki 2. satırdan sonraki tüm veriyi silelim:
with open("tahsilatlar2.txt","r+") as dosya: #dosyamızı okuma+yazma kipinde açtık.
    dosya.readline() #ilk satırın sonuna geldik.
    dosya.readline() #ikinci satırın sonuna geldik.
    kacbayt=dosya.tell() #imlecin bulunduğu yeri bayt olarak öğrendik ve bunu kacbayt değişkenine koyduk.
    dosya.truncate(kacbayt) #buradan itibaren budama yapalım.
    dosya.seek(0)#imleci başa getirelim
    print(dosya.read()) #baştan itibaren okuyalım.
```

```python
#Şimdi de dosyamızın boyutunu 10 kilobayta çıkaralım:
with open("tahsilatlar2.txt","r+") as dosya: #dosyamızı okuma+yazma kipinde açtık.
    dosya.truncate(1024*10)
    dosya.seek(0)
    print(dosya.read())
```

Dosya boyutuna baktığımız zaman gerçekten de 10 kb olduğunu görürüz

```python
#Son olarak dosyamı truncade() metodunu parametresiz kullandığımızda ne oluyor görelim:
with open("tahsilatlar2.txt","r+") as dosya: #dosyamızı okuma+yazma kipinde açtık.
    dosya.truncate()
    dosya.seek(0)
    print(dosya.read())
```

# İKİLİ (BİNARY) DOSYALAR

Basit metin dosyaları dışındaki tüm dosyalar (resim dosyaları, pdf dosyaları, world, excell dosyaları, video dosyaları...) pythonda binary dosyalardır. Aslında tüm dosyalar binarydir ancak python da metin dosyaları ile işlem yaparken binary özelliğine gerek yoktur.

```python
#Örnek bir pdf dosyasına bakalım:
dosya=open("ornek.pdf","br")#binary okuma kipinde açtık
```

```python
dosya.seek(0)
print(dosya.read(50))#ilk 50 baytını okuyalım
dosya.close()
```

Görüleceği üzere b harfi ile başlayan bir string var karşımızda. Burada en baştaki b harfi bize bu içeriğin bayt türünde olduğunu belirtiyor.Her bir \x ise kendinden sonra gelen sayının hexadecimal olduğunu belirtiyor.

```python
#pdf dosyalarında bulunan etiketlerden /Producer adındaki etiketi bulalım,
dosya=open("ornek.pdf","br") #dosyamızı binary okuma kipinde açtık
dosya.seek(0)
dosya_verisi=dosya.read() #tüm veriyi dosya_verisi stringine kaydettik
producer_index=dosya_verisi.index(b"/Producer") #dosya_verisi stringinin içinde "/Producer" stringinin başlangıç indexini
#producer_index adlı değişkene kaydettik
dosya.seek(producer_index)#imleci bu indexe getirdik
print(dosya.read(50))#bu indexten itibaren 50 bayt veriyi yazdırdık.
dosya.close()
#aradığımız üretici iText be lowagie.com
```
